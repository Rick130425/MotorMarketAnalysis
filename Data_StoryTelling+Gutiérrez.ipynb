## Limpieza y Transformación de Datos

Se eliminan las columnas dispensables para el análisis del mercado automotor. La columna de título es eliminada.

```python
df.drop('Title', axis=1, inplace=True)
```

Debido a que la cantidad de registros con valores nulos es relativamente pequeña en relación con la cantidad total de registros, se toma la opción de eliminarlos, en vez de realizar otra técnica, con el fin de evitar la modificación de distribución de los valores.

```python
df.dropna(inplace=True)
# También se eliminan las columnas que no cuenten con un valor útil
df = df.loc[(df != '-').all(axis=1)]
```

Se realiza la transformación de las columnas, con el objetivo de obtener los valores numéricos de las columnas, evitar duplicación de información entre columnas y facilitar el proceso de graficar los datos para el análisis exploratorio de datos (EDA).

```python
# Se renombra 'UsedOrNew' por 'Condition' (condición en inglés)
df = df.rename(columns={'UsedOrNew': 'Condition'})

# Se extrae la cilindrada del motor de la columna 'Engine'
df['Engine'] = df['Engine'].str.extract(r'(\d+(.\d+)?)\s*L', expand=False)[0]
# Se renombra 'Engine' por 'Displacement' (cilindrada en inglés)
df = df.rename(columns={'Engine': 'Displacement'})

# Se extrae el consumo de combustible (litros por cada 100 km)
df['FuelConsumption'] = df['FuelConsumption'].str.extract(r'(\d+(.\d+)?)\s*L', expand=False)[0]

# Se extrae el número de cilindros del motor
df['CylindersinEngine'] = df['CylindersinEngine'].str.extract(r'(\d+)')

# Se extrae el número de puertas
df['Doors'] = df['Doors'].str.extract(r'(\d+)')

# Se extrae el número de asientos
df['Seats'] = df['Seats'].str.extract(r'(\d+)')
```

Se asignan los tipos de datos correctos para cada columna.

```python
# Se establece la columna 'Year' como int
df['Year'] = df['Year'].astype('int')

# Se establece la columna 'Engine' como float
df['Displacement'] = df['Displacement'].astype('float')

# Se establece la columna 'FuelConsumption' como float
df['FuelConsumption'] = df['FuelConsumption'].astype('float')

# Se establece la columna 'Kilometres' como int
df['Kilometres'] = df['Kilometres'].astype('int')

# Se establece la columna 'CylindersinEngine' como int
df['CylindersinEngine'] = df['CylindersinEngine'].astype('int')

# Se establece la columna 'Doors' como int
df['Doors'] = df['Doors'].astype('int')

# Se establece la columna 'Seats' como int
df['Seats'] = df['Seats'].astype('int')

# Convierte la columna 'Price' en valores numéricos, y los que no se puedan
# convertir se establecen como NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
# Se eliminan todas las columnas con valores NaN
df.dropna(subset=['Price'], inplace=True)
```

Por último, se muestra la información del nuevo conjunto de datos.

```python
df.info()
```

# Modelos de Regresión

Realizada la etapa de observación de los datos, se continúa con la creación de modelos de regresión para predecir el coste de un auto en relación con sus características.

## Selección Inicial de Features Relevantes

Como primer punto se escogen las columnas que puedan ser útiles para el modelo. Principalmente, aquellas que tengan valores numéricos o que no tengan demasiados valores únicos de texto.

```python
df.info()
```

```python
df.nunique()
```

```python
columns_names = ['Brand', 'Year', 'Condition', 'Transmission', 'Displacement', 'DriveType', 'FuelType', 'FuelConsumption', 'Kilometres', 'CylindersinEngine', 'BodyType', 'Doors', 'Seats']
features = df[columns_names]
target = df['Price']
```

## Preprocesamiento de Datos

Después de la selección de factores, se realiza un preprocesamiento de los datos para que todas las características sean numéricas. Para las características expresadas en texto, se tomó la decisión de realizar un One Hot Encoding, debido a que no presentan un orden natural, como también no tienen una gran cantidad de valores únicos.

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

columns_to_encode = ['Brand','Condition', 'Transmission', 'DriveType', 'FuelType', 'BodyType']
encoded_data = encoder.fit_transform(features[columns_to_encode])

encoded_features = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_to_encode))

features = features.drop(columns_to_encode, axis=1).reset_index(drop=True)
encoded_features = encoded_features.reset_index(drop=True)
features = pd.concat([features, encoded_features], axis=1)
```

```python
features.info()
```

## Feature Selection

Se realiza la selección de características para los modelos de regresión. Para esto se utiliza la clase SFS del módulo feature_selection de scikit-learn.

```python
from sklearn.feature_selection import SequentialFeatureSelector  as SFS
from sklearn.linear_model import LinearRegression

sfs = SFS(LinearRegression(), n_features_to_select=0.20, direction='forward', cv=5, n_jobs=-1)

features_selected = sfs.fit_transform(features, target)
features.columns[sfs.get_support()]
```

## División de Dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_selected, target, test_size=0.2)
```

## Creación y Entrenamiento de Modelos

Se crean tres modelos: RandomForestRegressor, LinearRegression y Lasso, y se entrenan utilizando el conjunto de datos de entrenamiento.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Lasso

rf_model = RandomForestRegressor(n_estimators=600, max_features='sqrt', n_jobs=-1)
lr_model = LinearRegression(n_jobs=-1)
l_model = Lasso()

rf_model.fit(x_train, y_train)
lr_model.fit(x_train, y_train)
l_model.fit(x_train, y_train)
```

## Evaluación de Modelo

Se prueban los modelos usando el conjunto de prueba, y se presentan las métricas resultantes.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Debido a que la evaluación de los modelos se realiza de la misma manera para todos, se realiza una función de evaluación.

```python
def eval_model(model, x_train, x_test, y_train, y_test):
        train_y_pred = model.predict(x_train)
        test_y_pred = model.predict(x_test)

        # R2 score
        r2_train = r2_score(y_train, train_y_pred)
        r2_test = r2_score(y_test, test_y_pred)

        # Mean Squared Error (MSE) & Root Mean Squared Error (RMSE)
        mse_train = mean_squared_error(y_train, train_y_pred)
        mse_test = mean_squared_error(y_test, test_y_pred)
        rmse_train = np.sqrt(mse_train)
        rmse_test = np.sqrt(mse_test)

        # Mean Absolute Error (MAE)
        mae_train = mean_absolute_error(y_train, train_y_pred)
        mae_test = mean_absolute_error(y_test, test_y_pred)

        print(f"R2 (Train): {r2_train:.2f}, R2 (Test): {r2_test:.2f}")
        print(f"RMSE (Train): {rmse_train:.2f}, RMSE (Test): {rmse_test:.2f}")
        print(f"MAE (Train): {mae_train:.2f}, MAE (Test): {mae_test:.2f}")
```

```python
eval_model(rf_model, x_train, x_test, y_train, y_test)
```

```python
eval_model(lr_model, x_train, x_test, y_train, y_test)
```

```python
eval_model(l_model, x_train, x_test, y_train, y_test)
```

Se puede apreciar que el modelo RandomForestRegressor tiene un buen rendimiento tanto para los datos de entrenamiento como los datos de prueba, concluyendo que no se encuentra en overfitting ni en underfitting, y siendo el mejor modelo entre los tres.

Para el modelo LinearRegression y Lasso no se obtuvo un buen rendimiento en los datos de entrenamiento ni en los datos de prueba, llevando a la conclusión de que se encuentran en underfitting. Esto se puede deber a que son modelos menos complejos, en comparación al RandomForestRegressor.

Como conclusiones finales, probablemente sea posible mejorar su rendimiento reduciendo la cantidad de parámetros de entrada o modificando la complejidad de los mismos, lo cual puede ser la razón principal del underfitting.
